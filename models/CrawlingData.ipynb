```python
import pandas as pd
from datetime import datetime

# ... Kode sebelumnya ...

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
tweets_df = pd.read_csv(file_path, delimiter=";")

# Buat kamus untuk mengganti nama kolom
new_columns = {
    'conversation_id_str': 'Conv.ID',
    'tweet_url': 'URL',
    'created_at': 'Date',
    'full_text': 'Tweet',
    'id_str': 'ID',
    'reply_count': 'Replies',
    'retweet_count': 'Retweeted Tweet',
    'favorite_count': 'Likes',
    'lang': 'Language',
    'user_id_str': 'User ID',
    'username': 'Username'
}

# Ganti nama kolom
tweets_df = tweets_df.rename(columns=new_columns)

# Hapus kolom yang tidak diperlukan
tweets_df = tweets_df.drop(columns=['Conv.ID', 'Retweeted Tweet', 'URL', 'ID', 'Replies', 'Likes', 'Language', 'User ID'])

# Atur ulang urutan kolom
tweets_df = tweets_df[['Date', 'Username', 'Tweet']]

# Konversi format tanggal
tweets_df['Date'] = tweets_df['Date'].apply(lambda x: datetime.strptime(x, "%a %b %d %H:%M:%S +0000 %Y").strftime("%Y-%m-%d %H:%M:%S+00:00"))

# Tampilkan DataFrame
display(tweets_df)

# Simpan DataFrame yang telah diubah ke dalam file CSV di folder tweets-data-ready
output_path = f"../database/crawling/{filename}"  # Perhatikan perubahan path
tweets_df.to_csv(output_path, index=False)
```
